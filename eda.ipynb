```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv("/Users/priyankamalavade/Desktop/guvi_recommendation/cleaned_book_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Price"].fillna(df["Price"].median(), inplace=True)
```

```python
df=df.dropna(subset=['cleaned_description'])
```

```python
df.reset_index(drop=True, inplace=True)

```

```python
df = df.astype({
    'Rating': 'float64',
    'Number of Reviews': 'int64',
    'Price': 'int64',
    'Listening Time (minutes)': 'int64',
    'Rank': 'int64',
    'Book Name': 'object',
    'Author': 'object',
    'cleaned_description': 'object',
    'Genre': 'object'
})
```

```python
df.info()
```

```python
df.head()
```

```python
sns.set(style="whitegrid")
```

```python
# 1. Top Genres
plt.figure(figsize=(12, 6))
top_genres = df['Genre'].value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette="viridis")
plt.title('Top 10 Genres by Number of Books')
plt.xlabel('Number of Books')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
#2.Top 10 Most Frequent Authors
top_authors = df['Author'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_authors.values, y=top_authors.index, palette='viridis')
plt.title('Top 10 Most Frequent Authors')
plt.xlabel('Number of Books')
plt.ylabel('Author')
plt.tight_layout()
plt.show()
```

```python
#3. Distribution of Ratings

plt.figure(figsize=(10,6))
sns.histplot(df['Rating'], bins=20, kde=True, color='purple')
plt.title('Distribution of Book Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4. Relationship Between Rating and Price

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Price', y='Rating', hue='Genre', alpha=0.7)
plt.title('Rating vs Price')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

```

##  Rating vs Price Scatter Plot Analysis

This scatter plot visualizes the relationship between **book price** and **user rating**, with color indicating **genre**.

###  Observations

- Most audiobooks are **clustered at lower price ranges** (₹0–₹5000).
- Ratings are **consistently high (mostly 4 to 5)** across all price ranges, regardless of genre.
- Even **high-priced books** (above ₹10,000) do **not show lower ratings**, implying that **price does not negatively affect perception**.
- A few **low-rated books** exist, but they are not concentrated in any one genre or price band.
- Genres are **diverse across the price spectrum**, but most genres have a presence in the **lower price + higher rating zone**.

###  Insights

- There is **no strong correlation** between **Price and Rating**, confirming what was seen in the heatmap earlier.
- **High-priced books still maintain high ratings**, suggesting price may reflect **content depth or duration**, not necessarily popularity.
- Users seem to **value content quality**, not price – making **content-based recommendation systems** more effective than price-based ones.


```python
#5. Listening Time vs. Number of Reviews

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Listening Time (minutes)', y='Number of Reviews', alpha=0.7, color='darkorange')
plt.title('Listening Time vs Number of Reviews')
plt.xlabel('Listening Time (minutes)')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##  Listening Time vs Number of Reviews Scatter Plot

This scatter plot visualizes the relationship between the **length of audiobooks (in minutes)** and their **number of reviews**.

###  Observations

- A **major concentration** of audiobooks has a **listening time between 0 and 1000 minutes**.
- Within this range, some audiobooks have **very high numbers of reviews**, even **above 15,000**.
- After ~1000 minutes, the number of audiobooks decreases, and **most have fewer reviews**.
- There are a few **outliers** (long audiobooks with many reviews), but they are **rare**.

###  Insights

- **Short to medium-length audiobooks (100–600 min)** are the **most popular** in terms of user engagement (i.e., more reviews).
- There is **no direct linear correlation** between length and number of reviews – some shorter books still dominate in popularity.
- Users may prefer books that are **concise yet impactful**, contributing to more reviews.
- This can be helpful in **targeting preferred book durations** for recommendations or promotions.

###  Implications for Recommendation

- For users who prefer **popular books**, suggest audiobooks **under 700 minutes** with high review counts.
- For niche listeners, consider **longer audiobooks** in their genre of interest, even if reviews are lower.



```python
#6. Average Rating by Genre

avg_rating_genre = df.groupby('Genre')['Rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating_genre.values, y=avg_rating_genre.index, palette='coolwarm')
plt.title('Average Rating by Genre')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
#7.Listening Time Distribution 

plt.figure(figsize=(10,6))
sns.histplot(df['Listening Time (minutes)'], bins=30, kde=True, color='mediumseagreen')
plt.title('Distribution of Listening Time')
plt.xlabel('Listening Time (minutes)')
plt.ylabel('Number of Books')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#8. Correlation Heatmap 

plt.figure(figsize=(10,6))
numeric_cols = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_cols.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

##  Correlation Heatmap Analysis

A **correlation heatmap** visually represents the strength of relationships between numerical variables in the dataset. The correlation coefficient ranges from:

- **+1** → Perfect positive correlation  
- **0** → No correlation  
- **–1** → Perfect negative correlation  

In this dataset, we analyze the relationships among the following numeric features:

- `Rating`
- `Number of Reviews`
- `Price`
- `Listening Time (minutes)`
- `Rank`

###  Key Observations

| Feature Pair                        | Correlation | Interpretation |
|------------------------------------|-------------|----------------|
| `Rating` vs `Number of Reviews`    | **0.10**    | Weak positive correlation – books with more reviews may slightly tend to have higher ratings. |
| `Rating` vs `Price`                | **0.026**   | Negligible correlation – book price does **not significantly impact** ratings. |
| `Rating` vs `Listening Time`       | **0.019**   | No real relationship – longer books aren’t necessarily rated higher. |
| `Rating` vs `Rank`                 | **0.019**   | Very weak – popular (higher-ranked) books don’t always have better ratings. |
| `Price` vs `Listening Time`        | **0.16**    | Mild positive correlation – longer books tend to be **slightly** more expensive. |
| `Number of Reviews` vs `Listening Time` | **0.10**    | Slight tendency for longer books to receive more reviews. |

###  Insights

- **Ratings** appear to be independent of book price, length, or popularity rank.
- **Listening Time** shows a mild relationship with price and number of reviews.
- Overall, **no strong correlations** are observed between the features – indicating that user preferences might be influenced by other factors like **content or genre**.
